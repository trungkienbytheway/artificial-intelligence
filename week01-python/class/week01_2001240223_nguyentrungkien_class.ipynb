```python
import numpy as np
m = int(input("Nhập vào m dòng: "))
n = int(input("Nhập vào n cột: "))
a = np.random.rand(m, n) * 100
print(a)

k = int(input("Nhập vào dòng k: "))
print(a[k, :])

k = int(input("Nhập vào cột k: "))
print(a[:, k])

row_sums = np.sum(a, axis=1)
max_sum_index = np.argmax(row_sums)
max_sum = row_sums[max_sum_index]
largest_sum_row = a[max_sum_index, :]

print("Các tổng dòng: ", row_sums)
print("Tổng lớn nhất: ", max_sum)
print("Dòng có tổng lớn nhất: ", largest_sum_row)

tichcot = np.prod(a, axis=0)
index = np.argmin(tichcot)
tichnhonhat = tichcot[index]

print("Tích từng cột: ", tichcot)
print("Cột có tích nhỏ nhất là cột: ", index)
print("Tích nhỏ nhất là: ", tichnhonhat)

phantuchan = []
for i in range(0, a.shape[0], 2):
  for j in range(a.shape[1]):
    val = a[i, j]
    if val % 2 == 0:
      phantuchan.append(val)

if phantuchan:
  tbc = np.mean(phantuchan)
  print("Trung bình cộng: ", tbc)
else:
  print("Không có phần tử chẵn")

tong_bien = 0
dem_bien = 0
for i in range(m):
    for j in range(n):
        if i == 0 or i == m - 1 or j == 0 or j == n - 1:
            tong_bien += a[i, j]
            dem_bien += 1
if dem_bien > 0:
    print("Trung bình cộng các phần tử biên: ", tong_bien / dem_bien)

tich_trong = 1
dem_trong = 0
for i in range(1, m - 1):
    for j in range(1, n - 1):
        tich_trong *= a[i, j]
        dem_trong += 1
if dem_trong > 0:
    print("Trung bình tích các phần tử không thuộc biên: ", tich_trong ** (1/dem_trong))
else:
    print("Không có phần tử nằm bên trong biên.")
```

Bài 15

```python
import datetime
class SinhVien:
    def __init__(self, masv: str, ten: str, namsinh: int, diemtrungbinh: float):
        self.dssv = []
        self.masv = masv if len(masv) == 10 else "N/A"
        self.ten = ten if len(ten) <= 20 else "N/A"
        self.namsinh = namsinh
        self.diemtrungbinh = diemtrungbinh

    def __str__(self):
        return f"{self.masv} | {self.ten} | {self.namsinh} | {self.diemtrungbinh}"

    def nhapmang(self):
        n = int(input("Nhập vào n sinh viên: "))
        for i in range(n):
          print(f"Nhập thông tin cho sinh viên thứ {i + 1}")
          ms = input("Nhập mã số sinh viên: ")
          t = input("Nhập tên sinh viên: ")
          ns = int(input("Nhập năm sinh của sinh viên: "))
          dtb = float(input("Nhập điểm trung bình: "))
          sv = SinhVien(ms, t, ns, dtb)
          self.dssv.append(sv)

    def kiemtra_sinhvienlenlop(self):
        soluong_sinhvienlenlop = 0
        for sv in self.dssv:
          if sv.diemtrungbinh >= 5:
            soluong_sinhvienlenlop += 1
        return soluong_sinhvienlenlop

    def xuat_cacsinhviendu20tuoi(self):
      soluong_sinhviendu20tuoi = 0
      for sv in self.dssv:
        if (datetime.datetime.now().year - sv.namsinh) >= 20:
          soluong_sinhviendu20tuoi += 1
      print("Số lượng sinh viên đủ 20 tuổi", soluong_sinhviendu20tuoi)

    def dem_sosinhvien_hocdaihoc(self):
      dem = 0
      for sv in self.dssv:
        if (sv.masv[2:2] == "DH"):
          dem += 1
      return dem

    def sosinhvien_coten_Lan(self):
      sosinhvien = 0
      for sv in self.dssv:
        if "Lan" in sv.ten:
          sosinhvien += 1
      return sosinhvien

    def sosinhvien_coho_Phan(self):
      sosinhvien = 0
      for sv in self.dssv:
        if sv.ten.startswith("Phan"):
          sosinhvien += 1
      return sosinhvien

quan_ly = SinhVien("", "", 0, 0.0)
quan_ly.nhapmang()
quan_ly.xuat_cacsinhviendu20tuoi()
soluong_sinhvienlenlop = quan_ly.kiemtra_sinhvienlenlop()
print("Số lượng sinh viên lên lớp: ", soluong_sinhvienlenlop)
soluong_sinhvienhocdaihoc = quan_ly.dem_sosinhvien_hocdaihoc()
print("Số lượng sinh viên học đại học: ", soluong_sinhvienhocdaihoc)
sosinhvien_coten_Lan = quan_ly.sosinhvien_coten_Lan()
print("Số lượng sinh viên có tên Lan", sosinhvien_coten_Lan)
sosinhvien_coho_Phan = quan_ly.sosinhvien_coho_Phan()
print("Số lượng sinh viên có họ Phan", sosinhvien_coho_Phan)
```
