# **Bài 1**

```python
def tron_mang(a, b):
    c = []
    i = 0
    while i < len(a) and i < len(b):
        c.append(min(a[i], b[i]))
        i += 1
    c.extend(a[i:])
    c.extend(b[i:])
    return c


a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

print(tron_mang(a, b))
```

# **Bài 2**

```python
a = [
    [3, -2, 5],
    [4, 6, -1],
    [7, 8, 2]
]

def tong_tam_giac_tren_cheo_phu(a):
    n = len(a)
    s = 0
    for i in range(n):
        for j in range(n):
            if i + j <= n - 1:
                s += a[i][j]
    return s

def doi_am_thanh_duong(a):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] < 0:
                a[i][j] = abs(a[i][j])
    return a

def thay_chan_bang_x(a, x):
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i][j] % 2 == 0:
                a[i][j] = x
    return a

def toan_chan(a):
    for row in a:
        for x in row:
            if x % 2 != 0:
                return False
    return True

def doi_xung(a):
    n = len(a)
    for i in range(n):
        for j in range(n):
            if a[i][j] != a[j][i]:
                return False
    return True

def cheo_chinh_tang(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][i] > a[i + 1][i + 1]:
            return False
    return True

def tam_giac_duoi_cheo_phu(a):
    n = len(a)
    res = []
    for i in range(n):
        for j in range(n):
            if i + j >= n - 1:
                res.append(a[i][j])
    return res

def cheo_phu_giam(a):
    n = len(a)
    for i in range(n - 1):
        if a[i][n - 1 - i] < a[i + 1][n - 2 - i]:
            return False
    return True


print("Tong tam giac tren cheo phu:", tong_tam_giac_tren_cheo_phu(a))
print("Mang sau khi doi am thanh duong:", doi_am_thanh_duong([row[:] for row in a]))
print("Mang sau khi thay so chan bang 0:", thay_chan_bang_x([row[:] for row in a], 0))
print("Ma tran toan chan:", toan_chan(a))
print("Ma tran doi xung:", doi_xung(a))
print("Cheo chinh tang dan:", cheo_chinh_tang(a))
print("Tam giac duoi cheo phu:", tam_giac_duoi_cheo_phu(a))
print("Cheo phu giam dan:", cheo_phu_giam(a))
```
